# Load Data

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
file_path = '/kaggle/input/arageneval-train-data/train_data_splitted_topics.csv'

train_data = pd.read_csv(file_path)
display(train_data.head())
```

```python
file_path = '/kaggle/input/arageneval-train-data/valid_data_cleaned_topics.csv'

valid_data = pd.read_csv(file_path, index_col='Unnamed: 0')
display(valid_data.head())
```

```python
file_path = '/kaggle/input/arageneval-train-data/test_data_cleaned.csv'

test_data = pd.read_csv(file_path, index_col='Unnamed: 0')
display(test_data.head())
```

```python
# Select the relevant columns
train_data = train_data[['cleaned_text', 'topic_keywords', 'author']]
valid_data = valid_data[['cleaned_text', 'topic_keywords', 'author']]
test_data = test_data[['cleaned_text']]

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the author names to numerical labels
y_train_encoded = label_encoder.fit_transform(train_data['author'])
y_test_encoded = label_encoder.transform(valid_data['author'])


# Split data into features (X) and target (y)
X_train, X_test = train_data[['cleaned_text', 'topic_keywords']], valid_data[['cleaned_text', 'topic_keywords']]
y_train, y_test = y_train_encoded, y_test_encoded

print("Data splitting complete.")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

# Embed Text

```python
!pip install -qU sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import torch
```

```python
torch.cuda.empty_cache()
```

```python
embedding_model = SentenceTransformer("Omartificial-Intelligence-Space/GATE-AraBert-v1")
# embedding_model = SentenceTransformer("CAMeL-Lab/bert-base-arabic-camelbert-mix-pos-msa")
embedding_model[0].default_task = 'classification'

# Generate embeddings for training and testing data
X_train_text_embeddings = embedding_model.encode(X_train['cleaned_text'].tolist())
X_valid_text_embeddings = embedding_model.encode(X_test['cleaned_text'].tolist())
X_test_text_embeddings = embedding_model.encode(test_data['cleaned_text'].tolist())
X_train_keywords_embeddings = embedding_model.encode(X_train['topic_keywords'].tolist())
X_test_keywords_embeddings = embedding_model.encode(X_test['topic_keywords'].tolist())

# Combine text and keyword embeddings
X_train_embeddings = np.concatenate((X_train_text_embeddings, X_train_keywords_embeddings), axis=1)
X_test_embeddings = np.concatenate((X_test_text_embeddings, X_test_keywords_embeddings), axis=1)

print("Embedding generation complete using SentenceTransformer.")
print("X_train_embeddings shape:", X_train_embeddings.shape)
print("X_test_embeddings shape:", X_test_embeddings.shape)
print(type(X_train_embeddings))

# print("X_train_text_embeddings shape:", X_train_text_embeddings.shape)
# print("X_test_text_embeddings shape:", X_valid_text_embeddings.shape)
print("X_test_text_embeddings shape:", X_test_text_embeddings.shape)
# print(type(X_train_text_embeddings))
```

# Bagging with SVM

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC
from kaggle_secrets import UserSecretsClient
import wandb
from sklearn.utils import class_weight
from sklearn.metrics import classification_report, accuracy_score
```

## Training

```python
!pip install -qU wandb
```

```python
wandb.init(project="AraGenEval", name='svc_bagging_balanced_loss')
```

```python
# Balancing Class Weights

# Calculate class weights
class_weights = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)

# Create a dictionary of class weights
class_weight_dict = dict(zip(np.unique(y_train), class_weights))

print("Class weights calculated:")
print(class_weight_dict)
```

```python
# Initialize the base estimator (SVM)
# svm_base_estimator = SVC(probability=True, random_state=42, C=1.0, gamma='scale', kernel='rbf')
svm_base_estimator_weighted = SVC(
    probability=True,
    random_state=42,
    C=1.0,
    gamma='scale',
    kernel='rbf',
    class_weight=class_weight_dict
)


# Initialize the Bagging Classifier
# bagging_classifier = BaggingClassifier(
#     base_estimator=svm_base_estimator,
#     n_estimators=10, 
#     random_state=42,
#     n_jobs=-1
# )
bagging_classifier_weighted = BaggingClassifier(
    estimator=svm_base_estimator_weighted,  # Use the weighted SVM
    n_estimators=10,
    random_state=42,
    n_jobs=-1
)


# Log hyperparameters for BaggingClassifier and its base estimator (SVM)
# wandb.config.update({
#     "bagging_base_estimator": "SVC",
#     "bagging_n_estimators": bagging_classifier.n_estimators, 
#     "bagging_random_state": bagging_classifier.random_state, 
#     "bagging_n_jobs": bagging_classifier.n_jobs, 
#     "svm_probability": svm_base_estimator.probability, 
#     "svm_random_state": svm_base_estimator.random_state, 
#     "svm_C": svm_base_estimator.C, 
#     "svm_gamma": svm_base_estimator.gamma, 
#     "svm_kernel": svm_base_estimator.kernel, 
# })

# print("Wandb initialized and BaggingClassifier and SVM hyperparameters logged.")

# # Train the Bagging Classifier
# bagging_classifier_weighted.fit(X_train_text_embeddings, y_train)
bagging_classifier_weighted.fit(X_train_embeddings, y_train)

print("Bagging Classifier with SVM base estimator trained successfully.")
```

```python
# Make predictions on the train set
y_pred = bagging_classifier_weighted.predict(X_train_embeddings)

# Evaluate the model
accuracy = accuracy_score(y_train, y_pred)
report = classification_report(y_train, y_pred, output_dict=True) # Get report as dictionary

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report(y_train, y_pred)) # Print the standard report for readability

# Log metrics to wandb
# wandb.log({
#     "train_accuracy": accuracy,
#     "train_precision": report['macro avg']['precision'],
#     "train_recall": report['macro avg']['recall'],
#     "train_f1_score": report['macro avg']['f1-score']
# })

# print("Training metrics logged to wandb.")
```
