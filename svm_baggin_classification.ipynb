# Embed Text

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import torch
```

```python
torch.cuda.empty_cache()
```

```python
embedding_model = SentenceTransformer("Omartificial-Intelligence-Space/GATE-AraBert-v1")
embedding_model[0].default_task = 'classification'

# Generate embeddings for training and testing data
X_train_text_embeddings = embedding_model.encode(X_train['cleaned_text'].tolist())
X_test_text_embeddings = embedding_model.encode(X_test['cleaned_text'].tolist())
X_train_keywords_embeddings = embedding_model.encode(X_train['topic_keywords'].tolist())
X_test_keywords_embeddings = embedding_model.encode(X_test['topic_keywords'].tolist())

# Combine text and keyword embeddings
X_train_embeddings = np.concatenate((X_train_text_embeddings, X_train_keywords_embeddings), axis=1)
X_test_embeddings = np.concatenate((X_test_text_embeddings, X_test_keywords_embeddings), axis=1)

print("Embedding generation complete using SentenceTransformer.")
print("X_train_embeddings shape:", X_train_embeddings.shape)
print("X_test_embeddings shape:", X_test_embeddings.shape)
print(type(X_train_embeddings))
```

# Bagging with SVM

## Training

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
# Initialize the base estimator (SVM)
svm_base_estimator = SVC(probability=True, random_state=42)

# Initialize the Bagging Classifier
bagging_classifier = BaggingClassifier(
    base_estimator=svm_base_estimator,
    # n_estimators=10,
    random_state=42,
    n_jobs=-1 # Use all available cores
)
```

```python
param_grid = {'n_estimators': [10, 25, 50]} # Define a range of n_estimators to try

grid_search = GridSearchCV(estimator=bagging_classifier, param_grid=param_grid, cv=5, n_jobs=-1, verbose=3)

grid_search.fit(X_train_embeddings, y_train)

print("Best n_estimators:", grid_search.best_params_['n_estimators'])
```

## Validating

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Get the best model from the grid search
best_bagging_model = grid_search.best_estimator_

# Make predictions on the test set
y_pred = best_bagging_model.predict(X_test_embeddings)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(report)
```
